## Section 2: Configure Project with langgraph.json

A `langgraph.json` configuration file must be in your project root to tell LangGraph Studio where to find your graph.

````python
import json

# Your langgraph.json configuration (already created in project root)
config = {
    "dependencies": ["."],
    "graphs": {
        "agent": "./agent/graph.py:get_agent_graph"
    },
    "env": ".env"
}

print("langgraph.json configuration:")
print(json.dumps(config, indent=2))
print("\n✅ This file is already created in your project root at: /langgraph.json")
````

## Section 3: Set Environment Variables

Your `.env` file already has the required LangSmith configuration. Verify these are set:

````python
import os
from dotenv import load_dotenv

# Load environment variables from .env
load_dotenv()

# Required environment variables for LangSmith
env_vars = {
    "LANGSMITH_API_KEY": os.getenv("LANGSMITH_API_KEY"),
    "LANGSMITH_PROJECT": os.getenv("LANGSMITH_PROJECT"),
    "LANGCHAIN_TRACING_V2": os.getenv("LANGCHAIN_TRACING_V2", "true"),
    "LANGCHAIN_API_KEY": os.getenv("LANGCHAIN_API_KEY"),
}

print("✅ Environment Variables Check:")
for key, value in env_vars.items():
    status = "✓ SET" if value else "✗ NOT SET"
    print(f"{key}: {status}")
    
print("\n📝 Required .env entries:")
print("""
LANGSMITH_API_KEY=your-api-key-here
LANGSMITH_PROJECT=sandbox-agent
LANGCHAIN_TRACING_V2=true
LANGCHAIN_API_KEY=your-api-key-here
""")
````

## Section 4: Launch LangGraph Studio Locally

LangGraph Studio provides a web-based IDE for visualizing and debugging your graph in real-time.

````python
print("""
🚀 LAUNCH COMMANDS:

1. Basic launch (in your project root directory):
   $ langgraph dev

2. With explicit port:
   $ langgraph dev --port 8000

3. For Mac with Safari (tunnel mode for secure local access):
   $ langgraph dev --tunnel

📍 After running, the server starts at:
   http://localhost:8000  (or shown in terminal output)

💡 Features available in LangGraph Studio:
   ✓ Visualize graph nodes and connections
   ✓ Submit inputs and run the agent
   ✓ See step-by-step execution
   ✓ Inspect state at each step
   ✓ View LangSmith traces in real-time
   ✓ Debug tool calls and responses

⚠️ Mac Safari Note:
   If Safari blocks localhost connections, use the tunnel mode:
   $ langgraph dev --tunnel
""")
````

## Section 6: Alternative - Visualize with Mermaid in Jupyter

If you prefer not to use Docker or LangGraph Studio, you can generate a static Mermaid diagram of your graph directly in this notebook.

````python
import sys
sys.path.insert(0, '..')

# Import the graph
from agent.graph import get_agent_graph

# Get the compiled graph
graph, checkpointer = get_agent_graph()

# Display basic graph information
print("Graph Information:")
print(f"Graph name: {graph.get_graph().name}")
print(f"Graph type: {type(graph)}")
print("\nAvailable nodes:")

# Get graph structure
import inspect
graph_dict = graph.get_graph()
print(f"Nodes: {list(graph_dict.nodes.keys()) if hasattr(graph_dict, 'nodes') else 'N/A'}")
print(f"Edges: {list(graph_dict.edges) if hasattr(graph_dict, 'edges') else 'N/A'}")
````

````python
# Generate Mermaid diagram
from IPython.display import Markdown, display

graph_mermaid = graph.get_graph().draw_mermaid()

# Display the Mermaid diagram
display(Markdown(f"""
```mermaid
{graph_mermaid}
```
"""))

print("\n✅ Graph visualization generated using Mermaid!")
````
